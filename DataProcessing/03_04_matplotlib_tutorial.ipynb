### Figure, Axes

![Figure](https://storage.googleapis.com/skytowner_public/images/uG4nzrYfQ9KDk2M87YDI/matplotlib_figure_axes_axis%20(4).png)

```python
fig = plt.figure() # figure
ax1 = fig.add_subplot(2,2,1) # axes, 행의 개수, 열의 개수, 첫번째
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.plot(x, y)
ax1.set_title('Graph')

ax4.plot(x, y)

plt.show()
```

```python
x = [1, 3, 5]
y = [2, 8, 11]
z = [3, 4, 5]

plt.plot(x, y, 'b--', marker='*') # marker='o', '*'
# plt.plot(x, y, color='#00ff00', linewidth=5, linestyle='--') # color = 'r', 'g', 'b', 'pink', 'red'
plt.plot(x, z, color='navy') # 한 축 안에 여러 개의 그래프 그리기
```

범례 표시하기: [matplotlib_legend](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html)

```python
x = [1, 3, 5]
y = [2, 8, 11]
z = [3, 4, 5]

plt.plot(x, y, label='Eng')
plt.plot(x, z, label='Math')
plt.legend(loc='lower right') # 범례 추가, loc=위치지정
```

```python
fig = plt.figure(figsize=(10, 4)) # figure

ax1 = fig.add_subplot(1,2,1) # axes, 행의 개수, 열의 개수, 첫번째
ax1.scatter(x = df.loc[df['Species']=='Iris-setosa', 'SepalLengthCm'],
            y = df.loc[df['Species']=='Iris-setosa', 'SepalWidthCm'], color='r')
ax1.scatter(x = df.loc[df['Species']=='Iris-versicolor', 'SepalLengthCm'],
            y = df.loc[df['Species']=='Iris-versicolor', 'SepalWidthCm'], color='g')
ax1.scatter(x = df.loc[df['Species']=='Iris-virginica', 'SepalLengthCm'],
            y = df.loc[df['Species']=='Iris-virginica', 'SepalWidthCm'], color='b')
ax1.set_xlabel('SepalLengthCm')
ax1.set_ylabel('SepalWidthCm')
ax1.set_title('Sepal')

ax2 = fig.add_subplot(1,2,2)
ax2.scatter(x = df.loc[df['Species']=='Iris-setosa', 'PetalLengthCm'],
            y = df.loc[df['Species']=='Iris-setosa', 'PetalWidthCm'], color='r')
ax2.scatter(x = df.loc[df['Species']=='Iris-versicolor', 'PetalLengthCm'],
            y = df.loc[df['Species']=='Iris-versicolor', 'PetalWidthCm'], color='g')
ax2.scatter(x = df.loc[df['Species']=='Iris-virginica', 'PetalLengthCm'],
            y = df.loc[df['Species']=='Iris-virginica', 'PetalWidthCm'], color='b')
ax2.set_xlabel('PetalLengthCm')
ax2.set_ylabel('PetalWidthCm')
ax2.set_title('Petal')

plt.suptitle('iris scatter plot', fontsize=15)
plt.tight_layout()
plt.show()
```

```python
mydata = [[1, 2, 3, 4], 
          [10, 11, 12, 13], 
          [30, 31, 32, 33], 
          [41, 43, 45, 45]]

fig, ax = plt.subplots(2, 2)

ax[0, 0].plot(mydata[0])
ax[0, 1].plot(mydata[1])
ax[1, 0].plot(mydata[2])
ax[1, 1].plot(mydata[3])

plt.show()
```

```python
mydata = [[1, 2, 3, 4], 
          [10, 11, 12, 13], 
          [30, 31, 32, 33], 
          [41, 43, 45, 45]]

fig, ax = plt.subplots(2, 2)

for i in range(4):
    ax[i//2, i%2].plot(mydata[i])

plt.show()
```

```python
mydata = [[1, 2, 3, 4], 
          [10, 11, 12, 13], 
          [30, 31, 32, 33], 
          [41, 43, 45, 45]]

fig, ax = plt.subplots(2, 2)

for i, axis in enumerate(ax.flatten()):
    axis.plot(mydata[i])

plt.show()
```

```python
mydata = [[1, 2, 3, 4], 
          [10, 11, 12, 13], 
          [30, 31, 32, 33], 
          [41, 43, 45, 45]]

fig, ax = plt.subplots(2, 2)
ax = ax.reshape(-1)

for i in range(4):
    ax[i].plot(mydata[i])

plt.show()
```

```python
mydata = [[1, 2, 3, 4], 
          [10, 11, 12, 13], 
          [30, 31, 32, 33], 
          [41, 43, 45, 45]]

fig, ax = plt.subplots(2, 2)
ax = ax.reshape(-1)

for axis, z in zip(ax, mydata):
    axis.plot(z)

plt.show()
```

### Axes level

```python
plt.figure(figsize=(5,5))
sns.scatterplot(data=df, x='flipper_length_mm', y='bill_length_mm', hue='species')
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.histplot(data=df, x='species', hue='species')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.scatterplot(data=df, x='flipper_length_mm', y='bill_length_mm', hue='species', ax=axes[0])
sns.histplot(data=df, x='species', hue='species', ax=axes[1])
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 5))

for i, axis in enumerate(axes.flatten()):
    if i%2==0:
        sns.scatterplot(data=df, x='flipper_length_mm', y='bill_length_mm', hue='species', ax=axis)
    else:
        sns.histplot(data=df, x='species', hue='species', ax=axis)
plt.tight_layout()
```
