```python
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np

X_train = np.array([1.,2.,3.,4.,5.,6.])
y_train = np.array([9.,12.,15.,18.,21.,24.])

model = keras.Sequential([
    layers.Dense(units=1, input_shape=(1,)), # 완전 연결층(Fully Connected Layer), unit=출력의 개수
])
```

```python
model.summary() # None: batch 차원(데이터가 몇 개씩 들어올지 모름), 1: 출력의 개수,  Param #: W와 b의 개수(학습시킬 것)
```

```python
# 컴파일
model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mse', 'mae']
)
```

```python
model.fit(X_train,y_train,
          epochs=3000,
          batch_size=len(X_train),
          verbose=1)
```

```python
# 학습 파라미터
model.get_weights()
```

```python
# 평가
model.evaluate(X_train,y_train)
```

```python
X_test = [10.]

y_pred = model.predict(X_test)
y_pred
```
