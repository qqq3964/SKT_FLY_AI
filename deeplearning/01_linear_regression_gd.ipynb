```python
import numpy as np
import matplotlib.pyplot as plt
```

### 데이터준비

```python
# 3x+6 = y

X_train = np.array([1.,2.,3.,4.,5.,6.])
y_train = np.array([9.,12.,15.,18.,21.,24.])
```

```python
X_train
```

```python
plt.scatter(X_train,y_train)
plt.show()
```

### 학습
- 경사하강법(Gradient Descent)

```python
W = 0.0
b = 0.0
N = len(X_train)
nb_epochs = 10000
lr = 0.01

for epoch in range(nb_epochs) :
  # 가설
  hypothesis = W*X_train+b

  # cost
  cost = np.sum((hypothesis-y_train)**2)/N

  # 미분
  grad_W = 2*np.sum((hypothesis-y_train)*(X_train))/N
  grad_b = 2*np.sum((hypothesis-y_train))/N
  
  # update
  W = W-grad_W*lr
  b = b-grad_b*lr
  
  if epoch%200 == 0 :
    print("Epoch {:4d}, Cost:{:.10f} W : {:10f}, b : {:10f} ".format(epoch,cost,W,b))
W,b
```

```python
class LinearRegressor:
  def fit(self,X,y,epochs=100,lr=0.01):
    self.W_ = 0.0
    self.b_ = 0.0
    self.N_ = len(X)
    self.lr_ = lr
    for epoch in range(epochs) :
      # 가설
      hypothesis = self.W_*X+self.b_

      # cost
      cost = np.sum((hypothesis-y_train)**2)/self.N_

      # 미분
      grad_W = 2*np.sum((hypothesis-y_train)*(X))/self.N_
      grad_b = 2*np.sum((hypothesis-y_train))/self.N_
      
      # update
      self.W_ = self.W_-grad_W*self.lr_
      self.b_ = self.b_-grad_b*self.lr_
      
      if epoch%200 == 0 :
        print("Epoch {:4d}, Cost:{:.10f} W : {:10f}, b : {:10f} ".format(epoch,cost,self.W_,self.b_))
    
  def predict(self,X):
    return self.W_*X+self.b_
```

```python
LR = LinearRegressor()
LR.fit(X_train,y_train,epochs=5000)
```
