### 데이터 준비

```python
X_train = np.array([1.,2.,3.,4.,5.,6.])
y_train = np.array([9.,12.,15.,18.,21.,24.])
```

### 경사하강법

```python
# 초기화
W = tf.Variable(0.0)
b = tf.Variable(0.0)

y_hat = W*X_train+b
```

```python
# MSE
N = len(X_train)
cost = tf.reduce_mean(tf.square(y_hat-y_train))
```

```python
# 자동 미분
lr = 0.01

with tf.GradientTape() as tape :
  y_hat = W*X_train+b
  cost = tf.reduce_mean(tf.square(y_hat-y_train))

  W_grad,b_grad = tape.gradient(cost,[W,b])

  W.assign_sub(lr*W_grad)
  b.assign_sub(lr*b_grad)
W.numpy(),b.numpy()
```

```python
class LinearRegression:

  def fit(self,X,y,epochs=100,lr=0.01) :
    self.W = tf.Variable(0.0)
    self.b = tf.Variable(0.0)
    self.epochs = epochs
    self.lr = lr
    for epoch in range(self.epochs) :
      with tf.GradientTape() as tape :
        y_hat = self.W*X+self.b
        cost = tf.reduce_mean(tf.square(y_hat-y))

      W_grad,b_grad = tape.gradient(cost,[self.W,self.b])
      
      self.W.assign_sub(self.lr*W_grad)
      self.b.assign_sub(self.lr*b_grad)
      if epoch%200 == 0 :
        print("Epoch {:4d}, Cost:{:.10f} W : {:10f}, b : {:10f} ".format(epoch,cost.numpy(),self.W.numpy(),self.b.numpy()))
X_train = np.array([1.,2.,3.,4.,5.,6.])
y_train = np.array([9.,12.,15.,18.,21.,24.])   
LR = LinearRegression()
LR.fit(X_train,y_train,epochs=5000)
```

```python
# 초기화
W = tf.Variable(0.0)
b = tf.Variable(0.0)
epochs = 5000
lr = 0.01

for epoch in range(epochs):
  with tf.GradientTape() as tape:
    y_hat = W*X_train+b
    cost = tf.reduce_mean(tf.square(y_hat-y_train))

  W_grad,b_grad = tape.gradient(cost,[W,b])
  
  # W=W-lr*W_grad
  W.assign_sub(lr*W_grad)
  b.assign_sub(lr*b_grad)
  if epoch%200 == 0 :
    print("Epoch {:4d}, Cost:{:.10f} W : {:10f}, b : {:10f} ".format(epoch,cost.numpy(),W.numpy(),b.numpy()))
```
