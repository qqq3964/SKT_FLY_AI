```python
import os

data_root = '/content/cats_and_dogs_filtered'
data_root
```

```python
train_dir = os.path.join(data_root,'train')
val_dir = os.path.join(data_root,'validation')
```

### 파일명 리스트 획득

```python
# listdir를 모두 확인할수있다. list로 만들어줌 저경로 아래에 있는 모든 파일들의 이름들을 리스트로 만듬
train_cats_fname = os.listdir(os.path.join(train_dir,'cats'))
train_dogs_fname = os.listdir(os.path.join(train_dir,'dogs'))

print(f'cats {len(train_cats_fname)}')
print(f'dogs {len(train_dogs_fname)}')
```

```python
class_name = ['cat','dog']
```

```python
# PIL 을 이용해서 하드디스크에 있는 이미지를 메모리로 불러옴
import PIL

X_train = []
y_train = []

# 고양이

for i in range(len(train_cats_fname)) :
  # 1000개를 읽어 오겠다.
  f = os.path.join(train_dir,'cats',train_cats_fname[i])
  image = PIL.Image.open(f)
  # 모든 이미지가 다 같은게 아니니 비율이 다름 그래서 사진을 정사각형으로 만듬 옆으로 긴애는 정사각형으로 만들고 (나머지는 검은색으로 채움 opencv로)
  # 세로로 긴것도 마찬가지로 옆에 검은색으로 opencv로 채워넣음 화면이 안찌그러지도록 만듬
  image = image.resize((224,224))
  arr = np.array(image)
  X_train.append(arr)
  y_train.append(0)

for i in range(len(train_dogs_fname)) :
  f = os.path.join(train_dir,'dogs',train_dogs_fname[i])
  image = PIL.Image.open(f)
  image = image.resize((224,224))
  arr = np.array(image)
  X_train.append(arr)
  y_train.append(1)
```

# 검증데이터

```python
# 파일명 리스트 획득

val_cats_fname = os.listdir(os.path.join(val_dir,'cats'))
val_dogs_fname = os.listdir(os.path.join(val_dir,'dogs'))

X_val = []
y_val = []

# 고양이
for i in range(len(val_cats_fname)):
  f = os.path.join(val_dir,'cats',val_cats_fname[i])
  image = PIL.Image.open(f)
  image = image.resize((224,224))
  arr = np.array(image)
  X_val.append(arr)
  y_val.append(1)

# 강아지
for i in range(len(val_cats_fname)):
  f = os.path.join(val_dir,'cats',val_cats_fname[i])
  image = PIL.Image.open(f)
  image = image.resize((224,224))
  arr = np.array(image)
  X_val.append(arr)
  y_val.append(1)
```

```python
X_train = np.array(X_train)
y_train = np.array(y_train)
X_val = np.array(X_val)
y_val = np.array(y_val)
```
